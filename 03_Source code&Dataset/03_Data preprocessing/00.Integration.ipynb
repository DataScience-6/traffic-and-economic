```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Îç∞Ïù¥ÌÑ∞Î•º ÏùΩÏñ¥ÏòµÎãàÎã§
# traffic-> target value
# feature values
traffic=pd.read_csv("../data/a_pre/Seoul_traffic_rate_cleaned.csv",encoding="utf-8")

gdp=pd.read_csv("../data/a_pre/Gdp_cleaned.csv",encoding="utf-8")
gas_price=pd.read_csv("../data/a_pre/Gas_price_cleaned.csv",encoding="utf-8")
population=pd.read_csv("../data/a_pre/Population_seoul.csv",encoding="utf-8")
car_registration=pd.read_csv("../data/a_pre/Car_registration_cleaned.csv",encoding="utf-8")
min_wage=pd.read_csv("../data/a_pre/Min_wage_cleaned.csv",encoding="utf-8")
transition_fee=pd.read_csv("../data/a_pre/Trans_fee_cleaned.csv",encoding="utf-8")
consumer_price=pd.read_csv("../data/a_pre/Consumer_price_cleaned.csv",encoding="utf-8")
interest_rate=pd.read_csv("../data/a_pre/Interest_rate.csv",encoding="utf-8")
employed_num=pd.read_csv("../data/a_pre/Employed_num_cleaned.csv",encoding="utf-8")
unemployed_rate=pd.read_csv("../data/a_pre/Unemployed_rate_cleaned.csv",encoding="utf-8")
in_out=pd.read_csv("../data/a_pre/Import_and_Export_cleaned.csv",encoding="utf-8")
```

## GDPÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Îã§Î•∏ Îç∞Ïù¥ÌÑ∞Îì§Ïùò ÏÜçÏÑ±ÏùÑ Î∂ôÏó¨Ï§çÎãàÎã§

```python
df=gdp.copy()
df['Regular Gasoline']=gas_price['Regular Gasoline']
df['Diesel for automobiles']=gas_price['Diesel for automobiles']
df['population']=population['population']
df['Car registration']=car_registration['Car registration']
```

```python
df['Min wage']=min_wage['Min wage(hour)']
df['Min wage increase rate']=min_wage['Min wage Increase rate']

df['City bus trunk line']=transition_fee['City bus trunk line']
df['City bus wide area']=transition_fee['City bus wide area']
df['City bus circulation']=transition_fee['City bus circulation']
df['City bus late at night']=transition_fee['City bus late at night']
df['Town bus']=transition_fee['Town bus']
df['Subway']=transition_fee['Subway']
df['Texi']=transition_fee['Texi']
df['Model texi']=transition_fee['Model Texi']

df['Consumer price']=consumer_price['Consumer price index']

df['Interest rate']=interest_rate['Interest rate']

df['Number of employed']=employed_num['Number of employed(Seoul)']

df['Export amount']=in_out['Export amount']
df['Export growth rate']=in_out['Export growth rate']
df['Import amount']=in_out['Import amount']
df['Import growth rate']=in_out['Import growth rate']
df['Traffic rate']=traffic['Traffic rate']
df
```

```python
# Ï†ÑÏ≤òÎ¶¨ ÏûëÏóÖÏùÑ ÏúÑÌïú ÌòïÎ≥ÄÌôò. Ìïú Î≤àÎßå Ïã§Ìñâ Í∞ÄÎä•

# Date Ïª¨ÎüºÏùÑ datetime ÌòïÏãùÏúºÎ°ú Î≥ÄÌôò
df['date'] = pd.to_datetime(df['date'])

# 'population' Ïó¥Ïùò ÏâºÌëú Ï†úÍ±∞ Î∞è int64Î°ú Î≥ÄÌôò
df['population'] = df['population'].str.replace(',', '').astype(int)

#
df['Min wage'] = df['Min wage'].str.replace(',', '').astype(int)

df['Export amount'] = df['Export amount'].str.replace(',', '').astype(int)

df['Import amount'] = df['Import amount'].str.replace(',', '').astype(int)
```

```python
# Ìï¥Îãπ ÌååÏùºÏù¥ Ïó¥Î†§ÏûàÎäî Í≤ΩÏö∞ ÌååÏùº ÎßåÎì§Í∏∞Í∞Ä Í±∞Î∂ÄÎê®
#Ï†ÄÏû•ÌïòÍ∏∞
df.to_csv('../data/complete.csv', index=False)
df
```
