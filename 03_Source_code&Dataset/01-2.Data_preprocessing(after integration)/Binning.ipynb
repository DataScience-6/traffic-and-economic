```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dataest
data = pd.read_csv('../00_Data/Integrated_data/Fully_integrated_data.csv')
data.head()
```

```python
#  Define categories for economic growth
bins = [-float('inf'), 0, 2, 4, float('inf')]
labels = ['Negative Growth', 'Low Growth', 'Moderate Growth', 'High Growth']

# Dividing economic growth rates into categories
data['Economic Growth Rate'] = pd.cut(data['Economic Growth Rate (Real GDP Growth Rate)'], bins=bins, labels=labels)

#pd.cut : a function in the pandas library used to segment and sort data values into bins
# x : array-like input data
# bins : Can be an integer specifying the number of equal-width bins
data
```

```python
# 'Economic Growth Rate (Real GDP Growth Rate)' Remove a Column
data.drop('Economic Growth Rate (Real GDP Growth Rate)', axis=1, inplace=True)
```

```python
data
```

```python
# 'Check the basic statistics of 'Base Interest Rate'
data['Base Interest Rate'].describe()
```

```python
# Define categories for interest rates
#(categories are illustrated as examples and must be adapted for actual data)
bins = [0, 3.0, 5.0, float('inf')]
labels = ['Low', 'Moderate', 'High']

# Divide interest rates into categories
data['Interest Rate Category'] = pd.cut(data['Base Interest Rate'], bins=bins, labels=labels)

# Remove the original 'Base Interest Rate' column
data.drop('Base Interest Rate', axis=1, inplace=True)
```

```python
# Draw a histogram
plt.figure(figsize=(12, 6))

# Export price histogram
plt.subplot(1, 2, 1)
plt.hist(data['Export price'], bins=30, color='blue', alpha=0.7)
plt.title('Export Price Distribution')
plt.xlabel('Export Price')
plt.ylabel('Frequency')

# Import price histogram
plt.subplot(1, 2, 2)
plt.hist(data['Import price'], bins=30, color='green', alpha=0.7)
plt.title('Import Price Distribution')
plt.xlabel('Import Price')

plt.tight_layout()
plt.show()
```

```python
# EDivide the export price and import price into 
# four equal frequency intervals, respectively
data['Export Price Category'] = pd.qcut(data['Export price'], 4, labels=['Low', 'Medium', 'High', 'Very High'])
data['Import Price Category'] = pd.qcut(data['Import price'], 4, labels=['Low', 'Medium', 'High', 'Very High'])
# pd.qcut : uses quantiles to divide the data.
# This method ensures that each bin has roughly the same number of data points
# x: The one-dimensional input array or series to be binned.
# q: The number of quantiles. 
# labels: Optionally, labels for the bins
# Result Screen
print(data[['Export price', 'Export Price Category', 'Import price', 'Import Price Category']].head())
```

```python
#Remove the original price column
data.drop(['Export price', 'Import price'], axis=1, inplace=True)
```
