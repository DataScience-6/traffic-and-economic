```python
import featuretools as ft
import pandas as pd
```

```python
data=pd.read_csv("../data/integration_b.csv",encoding="utf-8")
data.head()
```

```python
# Create an entity set
es = ft.EntitySet(id='economic_data')

# Add the data as an entity
es = es.add_dataframe(dataframe_name='data', dataframe=data, index='index', make_index=True)

# Run deep feature synthesis (DFS) to create new features
feature_matrix, feature_defs = ft.dfs(
    entityset=es,
    target_dataframe_name='data',
    agg_primitives=['mean', 'sum', 'min', 'max', 'std','count'],
    trans_primitives=['month', 'year', 'diff'],
    max_depth=5) 

feature_matrix.head()
```

```python
print([f.get_name() for f in feature_defs])
```
